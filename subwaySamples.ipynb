```python
queens = reg_theory.loc[(reg_theory["neighbourhood_group"]=="Queens")].sample(n=3)
queens
```

```python
bronx = reg_theory.loc[(reg_theory["neighbourhood_group"]=="Bronx")].sample(n=3)
bronx
```

```python
statenIsland = reg_theory.loc[(reg_theory["neighbourhood_group"]=="Staten Island")].sample(n=3)
statenIsland
```
